```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.countplot(x='gender',data=df)
plt.title("Number of students by gender")
plt.show()
```

```python
sns.histplot(df['math score'],bins=20,kde=True)
plt.title("distribution of maths scores")
plt.show()
```

```python
sns.histplot(df['reading score'],bins=20,kde=True)
plt.title("distribution of reading scores")
plt.show()
```

```python
sns.histplot(df['writing score'],bins=20,kde=True)
plt.title("distribution of writing scores")
plt.show()
```

```python
sns.scatterplot(x='math score',y='reading score',data=df)
plt.title("maths score vs reading score")
plt.show()
```

```python
sns.scatterplot(x='math score',y='writing score',data=df)
plt.title("maths score vs writing score")
plt.show()
```

```python
sns.scatterplot(x='writing score',y='reading score',data=df)
plt.title("writing score vs reading score")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
df=pd.read_csv("../data/StudentsPerformance (1).csv")
df.head()
plt.figure(figsize=(6,4))
sns.heatmap(df.corr(numeric_only=True),annot=True,cmap="coolwarm")
plt.title("Correlation between subjects")
plt.show()
```
